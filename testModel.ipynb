```python
import cv2 as cv
from ultralytics import YOLO

model_path = 'output/yolo_model/weights/best.pt'

model = YOLO(model_path)

# Open the video file
video_path = "./data/video.mp4"
cap = cv.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()
    
fourcc = cv.VideoWriter_fourcc(*'mp4v')
out = cv.VideoWriter('output_video.mp4', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

    
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    results = model(frame)
    
    for result in results:
        frame = result.plot()
    
    # Write the frame to the output video
    out.write(frame)
    
cap.release()
out.release()
cv.destroyAllWindows()
```
